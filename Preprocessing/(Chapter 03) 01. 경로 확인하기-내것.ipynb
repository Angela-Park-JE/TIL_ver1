## 경로 설정하기

파이썬에서 데이터를 불러오려면 반드시 경로와 확장자까지 포함시켜서 불러와야 한다.  

> 경로를 설정하여, 파일을 불러올 때마다 경로를 포함시키는 번거로움을 해결한다.  
OS 모듈에 속한 함수를 사용하면 경로를 확인하고 설정할 수 있다.  
`os.getcwd()` : 현재 경로를 반환  
`os.chdir(path)` : 현재 경로를 path로 지정
