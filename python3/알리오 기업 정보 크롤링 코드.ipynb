# 알리오 크롤링 최종

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def crawl_table(url):
    # 페이지에 접속하여 HTML 가져오기
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # class가 "tbl type_03"인 테이블 찾기
    table = soup.find('table', class_='tbl type_03')
    
    # 테이블에서 각 행(tr) 찾기
    rows = table.find_all('tr')
    
    # 데이터를 담을 리스트 초기화
    data = []

    # 테이블의 각 행(tr)에 대해 반복적으로 데이터 추출
    for row in rows:
        columns = row.find_all('td')
        if columns:
            head = columns[1].text.strip()
            body1 = columns[2].text.strip()
            body2 = columns[3].text.strip()
            body3 = columns[4].text.strip()
            data.append([head, body1, body2, body3])

    # 데이터프레임 생성
    df = pd.DataFrame(data, columns=['head', 'body1', 'body2', 'body3'])
    
    return df


# 빈 데이터 테이블
data = pd.DataFrame(columns = ['head', 'body1', 'body2', 'body3'])
# 현재는 페이지가 35 까지 있음 (24.02.28)
for i in range(1, 36):
    url = f"https://job.alio.go.kr/orginfo.do?pageNo={i}&apba_id=&jidt_dptm=&apba_na="
    result_df = crawl_table(url)
    data = pd.concat([data, result_df], ignore_index = True, sort = False)

# 데이터 확인
data.head(10)
```
