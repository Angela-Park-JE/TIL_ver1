ì´ë•Œ ìì¹´ë“œì´ê¸° ë•Œë¬¸ì— `linkage` ëŠ” `ward`ë¥¼ ì“¸ ìˆ˜ ì—†ê¸° ë•Œë¬¸ì— `average`ë¥¼ ì‚¬ìš©í•œë‹¤.

```python
# êµ°ì§‘í™” ëª¨ë¸ ì¸ìŠ¤í„´ìŠ¤í™” ë° í•™ìŠµ
from sklearn.cluster import AgglomerativeClustering as AC
clustering_model = AC(n_clusters = 10,
                      affinity = "jaccard",
                      linkage = "average")
clustering_model.fit(matrix_df)
```

### ë¼ë²¨ ì¶”ì  ë° ë°ì´í„°í”„ë ˆì„ ìƒì„±

ë³´ê¸° ì¢‹ê²Œ ë°ì´í„° í”„ë ˆì„ìœ¼ë¡œ ë°”ê¾¸ëŠ”ë°, íšŒì› IDë¥¼ ë°”íƒ•ìœ¼ë¡œ ì†Œì† êµ°ì§‘ì„ ê°’ìœ¼ë¡œ ê°–ëŠ” ë°ì´í„° í”„ë ˆì„ì„ ìƒˆë¡œ ë§Œë“ ë‹¤. 

```python
cluster_labels = clustering_model.labels_
cluster_labels
```

```python
# íšŒì›ë³„ ì†Œì† êµ°ì§‘ ì…ë ¥ ë° í™•ì¸
cluster_info = pd.DataFrame({"íšŒì›ID":matrix_df.index, 
                             "ì†Œì†êµ°ì§‘":cluster_labels})
cluster_info.head()
```
