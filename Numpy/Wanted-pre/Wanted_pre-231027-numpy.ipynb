## 2. 서로 다른 형태의 Numpy 연산
- Numpy는 서로 다른 형태의 배열을 연산할 때는 행 우선으로 수행됩니다.

```python
array1 = np.arange(4).reshape(2, 2) # 2by2
array2 = np.arange(2) # 1by2
print(array1)
print(array2)
```

```python
array1 + array2
```

- 브로드캐스팅 적용 예시2

```python
array1 = np.arange(0, 8).reshape(2, 4)
array2 = np.arange(0, 8).reshape(2, 4)
# 두 어레이 합치기 - 세로축으로
array3 = np.concatenate([array1, array2], axis = 0)

print(array3)
```

```python
array4 = np.arange(0, 4).reshape(4, 1) # 4X 1
print(array4)
```

```python
print(array3+array4)
```

## 3. Numpy의 마스킹 연산

```python
array1 = np.arange(16).reshape(4, 4)
print(array1)
```

```python
# 10보다 큰가?
array2 = array1 > 10
print(array2)
```

```python
# 마스킹 연산의 활용
# TF마스킹된 배열을 넣게되면
# 그 조건을 만족하는 데이터만 어떠한 작업을 하도록 명시할 수 있다.

# 조건에 맞는 원소들만 값이 바뀌도록 함

array1[array2] = 100
print(array1)
```

```python
# 이 작업은 조건에 맞는 원소들의 배열이 아니라 그 원소들의 값만 바꾼다.

array1 = np.arange(16).reshape(4, 4)
array2 = array1 > 10

array1[array2] * 100  # 10초과의 값들이 100 씩 곱해짐
```

## 4. Numpy의 집계함수



```python
import numpy as np

array = np.arange(16).reshape(4, 4)

print(array)
print('최대값:', np.max(array))
print('최소값:', np.min(array))
print('합계:', np.sum(array))
print('평균값:', np.mean(array))
```

각각의 축에 대해서 값을 구하도록 수행할 수도 있다.

```python
print(array)
print('최대값:', np.max(array, axis = 0))
print('최소값:', np.min(array, axis = 0))
print('합계:', np.sum(array, axis = 0))
print('평균값:', np.mean(array, axis = 0))
```
