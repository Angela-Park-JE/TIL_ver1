# 한 계정에서 동시 접속이 가능한가?

```python
# 한 계정안의 online 여부 개수

df[['account', 'name', 'id', 'online']]
```

```python
# 상태가 두 가지 이상인 캐릭터들은 1 이상이 True이다

simulconnect_account = df.groupby(['account'], observed = False)['online'].sum() > 1
simulconnect_account.sample(10)
```

위의 simulconnect_account는 account 별로 online들의 값들 즉 True값 합이 1초과인 경우를 가져온다.

한 account 내 online의 True가 한개인 경우는 한 account 내에서 여러 캐릭터들 중 한 캐릭터를 정상적으로 접속한 것으로 볼 수 있다는 점을 생각하여

합이 1 초과이면 한 account에서 여러 캐릭터를 접속한 상태라고 볼수 있다.

그렇다면 누가 여러 캐릭터를 동시 접속한 상태인 것일까?

```python
simulconnect_account[simulconnect_account == True]
```

```python
simulusers= simulconnect_account[simulconnect_account == True].index
simulusers
```

```python
# 위의 캐릭터들을 데이터 셋에서 가져온다.

for i in simulusers:
    arg = df['account'] == i
    Frame = df[arg][['account', 'name', 'id', 'online']]
    print(Frame)
```

```python
blacklist_users = list(simulusers)
# bin_df = pd.DataFrame()

for u in range(len(blacklist_users)):
    acc = blacklist_users[u]
    if u == 0:
        bin_df = df[df['account'] == acc]
    elif u != 0:
        bin_df2 = df[df['account'] == acc]
        whole_df = pd.concat([bin_df, bin_df2], axis = 0)
        bin_df = whole_df
```

```python
bin_df
```

한 유저가 동시에 접속이 가능한가 본 것인데, 실제로 한 유저 안에서 online이 두개인 캐릭터들이 존재했다. 

존재했다는 것만으로 이상함을 느끼는데, 실제로 문제가 되는 게임규정이라면 이 유저들을 상태로 검색해 볼 수 있을 것 같다.

## by one twitch account

익숙한 직업을 할까? 그들은 특정 직업에만 능통한 편일까? 아니면 다 잘하는 편일까?

먼저 한 트위치 계정을 한 사람으로 가정하고 그들이 선택한 직업들을 보기로 한다.

```python
grouped_twitch = df[['account','name', 'id', 'twitch']].groupby(by = 'twitch', as_index = 'twitch').nunique().sort_values('id', ascending = False)
grouped_twitch
```

```python
# 한 개 이상 아이디를 가진 트위치 유저들. 그 중 다 계정을 사용중인 사람들부터 쉽게 보기위해 정렬.

grouped_twitch[grouped_twitch['id']>1].sort_values('account', ascending = False, ignore_index = False)
```

```python
# 인덱싱 다시

bin_df = grouped_twitch[grouped_twitch['id']>1].sort_values('account', ascending = False).reset_index()
bin_df
```

```python
# merge를 이용하여 각 유저들이 가진 직업들이나 모드 등을 데려온다

twitch_df = pd.merge(bin_df, df[['twitch', 'class', 'stat_Str', 'stat_Dex', 'stat_Int', 'ladder']])
twitch_df
```

```python
twitch_df.groupby(by = 'twitch', as_index = False)['class'].nunique()
```

```python
twitch_df.groupby(by = 'twitch', as_index = False)['class'].nunique()['class'].mean()
```

### Commentary

한 트위치 아이디가 보통 두 세가지의 직업을 하는 것으로 나타났다.
