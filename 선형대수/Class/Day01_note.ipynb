## 벡터의 덧셈을 하는 함수 만들기

```python
def v_add(u,v):    
    '''
    벡터의 덧셈
    입력값: 더하고자 하는 벡터 u,v
    출력값: 벡터 u, v 의 더한 결과 w
    '''
    n = len(u)
    w = []
    
    for i in range(0, n):
        val = u[i] + v[i]
        w.append(val)
        
    return w
```

```python
u = [1, 2, 3]
v = [4, 5, 6]
v_add(u,v)
```

# pandas 없이 해보기

```python
import csv
```

```python
raw = []

with open('./data/mall_customer.csv', 'r', encoding = 'utf-8') as f:
    rdr = csv.reader(f)
    for line in rdr:
        raw.append(line)
```

csv 함수에 reader라는 함수가 있다. with open이라는 것의 용법.

행이 리스트 형태로 들어가 있다. 개발을 할 때는 행 중심이라는 것이 그 말.

그리고 작은 따옴표로 들어간 것, 전부 object로 들어가 있는 상황이라 숫자로 변환해주어야 한다.

```python
raw
```

## 행렬 살피기

```python
n_row = len(raw)
n_col = len(raw[0])
```

```python
print(n_row)
print(n_col)
```

데이터를 불러왔으면 크기를 확인하는 습관을 들여야 한다.

201로 1이 더 나온 이유는 컬럼 행 때문에 더 나온 것.

## 기존 데이터를 실수형태로 바꾸기

```python
data = []
for i in range(1, n_row):     # 0번 행은 제목 행이라 안쓴다.
    row = []
    for j in range(0, n_col):
        val = float(raw[i][j])
        row.append(val)
    data.append(row)
```

```python
data
```

```python
data[2][2]
```

```python
data[2][2] + data[3][2]
```

## 벡터 뽑기

```python
age = [] 

for i in range(0, n_row-1):  # 한 줄 씩 조회하기.
    val = data[i][2]         # 2가 Age 열의 데이터이기 때문에.
    age.append(val)
```

```python
age
```

pandas에는 시리즈라는 자료형으로 나올 것이다. 시리즈와 데이터프레임끼리 연산을 못한다던가 문제가 생긴다. 

그러나 우리가 지금 가져온 것은 리스트라서 연산이 안될 수가 없다. 판다스로 하다가 안될 때 바꿔서 하면 문제 해결! ㅎㅁㅎ

```python
score = []

for i in range(0, n_row-1):
    val = data[i][4]
    score.append(val)
```

```python
score
```

pandas로 하지만, 층이 하나씩 올라갈 수록 위험성도 높아지는 것을 인지하는 것이 좋다. 쓸 수 있으면 좋지만, 못쓰는 상황이 올 수도 있다는 것을 생각해야.

```python
res = np.add(age, score)
```

```python
res
```

우리가 리스트와 리스트 즉 벡터와 벡터를 넣었더니 결과가 바로 나온다는 것이에요.

```python
type(res)
```

두 개를 더했더니 numpy.ndarray 라는 자료형이라는 것이다. 

darray라는 자료형은 numpy의 자료형이다. 이대로 작업하면 또 이것떄문에 문제가 생길 수 있다.
