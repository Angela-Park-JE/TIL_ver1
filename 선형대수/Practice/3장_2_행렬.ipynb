# 행렬

## 행렬의 정의

행렬은 사각형 형태로 숫자를 나열하는 것을 의미한다. 

행렬은 row와 column으로 구성되어 있으며 각각 행은 가로 방향, 열은 세로 방향을 의미한다.



*이미지 출처: https://ko.wikipedia.org/wiki/%ED%96%89%EB%A0%AC*

행렬은 앞서 배운 벡터의 집합이다. 행 벡터의 집합이라고 볼 수도 있고, 열 벡터의 집합이라고 볼 수도 있다.

n by p 행렬이라고 할 때, n개의 행 벡터의 집합, p개의 열 벡터의 집합이라고 생각할 수 있다.

대문자로 나타내고, 원소를 스칼라 형태로 쓰고, 오른쪽 아래에 행 번호, 열 번호를 순서대로 표기한다.

행 크기와 열 크기가 동일하다면 해당 행렬을 정사각 행렬(square matrix)라고 부른다.

## 행렬의 덧셈과 뺄셈

각 행렬의 동일 위치에 대응하는 원소끼리 덧셈 혹은 뺼셈을 한다. 

가능하려면, 연산하려는 행렬의 크기가 동일하다는 전제가 필요하다.

동일 위치 : 원소의 행 번호와 열 번호가 동일하다는 뜻



### 파이썬 구현

```python
A = [[2, 7], [3, 4], [6, 1]]
B = [[1, 4], [4, -1], [2, 5]]

n = len(A)    # 행의 크기
p = len(A[0]) # 열의 크기

res = []
for i in range(0, n): # i번째 행
    row = []
    for j in range(0, p): # j번째 열
        val = A[i][j] + B[i][j] # 같은 위치에 있는 원소끼리의 합
        row.append(val)
    res.append(row)
```

```python
print(res)
```

'add' 라는 함수로 구현하기

```python
def add(A, B):
    """
    행렬의 덧셈
    입력값: 행렬의 덧셈을 수행할 행렬 A, B
    출력값: 행렬 A와 B의 덧셈 결과인 C
    """
    n = len(A)
    p = len(A[0])
    
    res = []
    for i in range(0, n):
        row = []
        for j in range(0, p):
            val = A[i][j] + B[i][j]
            row.append(val)
        res.append(row)
    
    return res
```

```python
# 시험

A = [[2, 7], [3, 4], [6, 1]]
B = [[1, 4], [4, -1], [2, 5]]

C = add(A, B)
print(C)
```
