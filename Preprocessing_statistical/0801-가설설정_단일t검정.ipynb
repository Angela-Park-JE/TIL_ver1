## 통계적 가설 검정
데이터를 바탕으로 통계적으로 가설의 유의미 여부를 검정하는 것을 통계적 가설 검정이라 한다.  
**가지고 있는 샘플**을 가지고 **모집단의 특성**에 대한 가설에 대한 통계적 유의성을 검정하는 일련의 과정.  
가지고 있는 데이터로 전체 데이터(모집단)을 알고 싶은 상태이다. 모집단을 정확히는 알 수 없다. 그래서 모집단의 부분집합인 가지고 있는 데이터로 추정하는 과정을 거치게 된다.  
- 수집한 데이터는 매울 특별한 경우를 제외하고는 전부 샘플이며, 모집단은 정확히 알 수 없는 경우가 더 많음  
- 통계적 유의성: 어떤 실험 결과(데이터)가 확률적으로 봐서 단순한 우연이 아니라고 판단될 정도로 의미가 있음을 뜻함. 즉 단순한 우연이라고 치부하기엔 확률이 작다, 우연이 아니다는 의미. 

통계적 가설 검정과정은 5단계이다
- 영 가설과 대립 가설 설정  
- 가설을 검정하기 위한 검정 통계량 설정  
- 가설의 유의성을 판단하기 위한 기각역 설정  
- 검정 통계량 계산  
- 가설이 유의한지 아닌지 의사 결정  



## 영 가설(귀무 가설)과 대립 가설
##### 영 가설(null hypothesis, $H_0$)
> 정의  
    - 특졀한 증거가 없으면 참으로 추정되는 가설  
    - 우리의 관심 대상이 아닌 가설(검정으로 영 가설을 기각해야 통계적 의미가 있음으로 판단)  
    - 기각하기 전까지는 참이라고 봄  
> 설정 방법  
    - 모집단에 대한 특성을 등호로 표기

##### 대립 가설(alternative hypothesis, $H_1, H_a$)
> 정의  
    - 특별한 증거가 없으면 거짓으로 추정되는 가설  
    - 우리의 관심 대상인 가설  
> 설정 방법  
    - 모집단에 대한 특성을 부등호로 표기(단측 검정, 양측 검정)  
> 양측 검정  
    - $H_0$ : n과 같다.  
    - $H_1$ : n과 같지 않다. (미만이거나 초과일 경우 둘 다 해당)  
> 단측 검정  
    - $H_0$ : n와 같다.  
    - $H_1$ : n보다 크다.  

## 오류의 구분

가설 검정에서 발생하는 오류는 두 가지로 나타난다. 귀무가설을 기준으로 참을 거짓으로, 거짓을 참으로 하여 1종과 2종으로 나눈다.  

> 1종 오류(type 1 Error)  
: 참을 거짓으로 하는 것. 참인데도 불구하고 귀무 가설이 가짜라고 하는 오류.   

> 2종 오류(type 2 Error)  
: 거짓을 참으로 하는 것. 귀무 가설이 거짓인데 참이라고 하는 오류.  


## 유의 확률, p-value

유의 확률 p-value: '영 가설이 맞다고 가정할 때 얻은 결과'와 다른 결과가 관측될 확률로, 그 값이 작을 수록 영 가설을 기각할 근거가 된다.  
보통, **p-value가 0.05 혹은 0.01 미만이면 영 가설을 기각한다.**  

예를 들어, 영 가설이 맞다고 가정하고, 데이터 관측을 했더니, 영 가설이 참일 때 나올 수 있는 데이터가 아니었다. 그럼 영 가설을 기각할 수 있다.  

주의! 영 가설을 기각한다고 해서 반드시 대립 가설이 참인 것은 아니다.  특히 단측 검정에서 그렇다.

## 단일 표본 t검정
그룹의 평균이 기준 값과 차이가 있는지 확인하는 검정 방법  
어떠한 가설 검정을 공부할 때는 목적을 제대로 안 다음, 그것의 영 가설과 대립 가설을 아는 것이 중요하다. (! 그리고 구현하는 것은 어렵지 않고, 가설이 무엇이고, 검정할 때 이 방법을 쓰면 좋겠다고 생각하게 되는 것이다.)

**하나의 영 가설에도 여러개의 대립가설이 생길 수 있다**.  
$H_0$ : $\bar {x}  = \mu$, 평균이 기준 값과 같다.    
$H_1$ : 같지 않다, 기준 값보다 크거나, 작거나.  

### 단일 표본 t-검정의 선행 조건
- 정규 분포를 따라야 수행할 수 있음.  
    따라서 Kolmogorov-Smornov나 Shapiro-Wilk를 사용한 **정규성 검정**이 선행되어야 한다.  
- 일반적으로 데이터가 30개 이상이면 정규성을 띌 것으로 보아도 무방할 정도이다. 

- 그렇지만 샘플 수가 많을 수록 정규성을 띌 가능성이 높아지므로, 샘플 수가 부족한 경우에만 정규성 검정을 수행한 뒤, 정규성을 띄지 않는다고 판단되면, **비모수적 방법인 부호 검정(sign test)**나 **윌콕슨 부호 - 순위 검정**을 수행해야 한다.(구현 방법만 알아보기로 함)  


### 단일 표본 t-검정 통계량

분자에서, 표본평균과 기준 값의 차이가 크면 클 수록 통계량의 절대값이 커진다는 의미. 분모에서, 표준편차가 클 수록 t의 절대값을 줄어들게 한다. 표본수가 많을수록 t는 커진다. 다시말해, 표본평균과 기준 값의 차이가 크면 클 수록 통계량의 절대값이 커질 것이고, 통계량의 절대값이 크면 클수록 유의 확률이 줄어드는 구조이다.  
다시 말해, **표본 평균과 기준 값의 차이가 크고, 표준편차가 작고 표본수가 많을 수록 영 가설이 기각될 확률이 높아진다.**  
= 표본 데이터와 귀무 가설의 차이가 증가할수록 t라는 절대값도 증가한다.

정규분포와 유사한 t분포 상에 위치시키는 방식으로 p-value를 계산하게 된다.  
p-value는 대립가설에 따라 조금씩 다르게 계산한다.  

만약 $ \bar {x} > \mu $ 혹은 $ \bar {x} < \mu $ 의 단측검정일 경우에는 t의 위치가 각각 오른쪽과 왼쪽일 것이고, p-value는 t를 기준으로 바깥이 될 것이다. 즉, p-value가 각각 오른쪽(p-value가 t보다 크면 대립 가설이 참이 된다), 왼쪽에 위치하고(p-value가 t보다 작으면 대립 가설이 참이 된다),  
$ \bar {x} \not= \mu$ 인 양측 검정일 경우는 각 양 꼬리에 $t$ 와 $-t$가 위치하면서 각각 p-value /2 가된다. 


### 정규성 검정 : Kolmogorov - Smornov(ks test)
KS test는 관측한 샘플들이 특정 분포를 따르는지 확인하기 위한 방법이다.  
정규분포를 따르는지 검정하는 것이라기 보다, 관측한 샘플들이 어떤 분포를 따르는지 검정하는 방법이며, 특정 분포를 정규 분포로 설정하면 정규성 검정에도 쓰일 수 있다.  

우리가 가진 실제 값과, 특정 분포를 따른다면 나올 것이라 예상되는 값의 차이가 크면 클 수록 특정 분포를 따르지 않는다고 볼 수 있다. 만약 우리가 가진 값이 특정 분포를 따른다면, 그 차이는 줄어든다.  


## 파이썬을 이용한 단일 표본 t검정

- 정규성 검정(KS test)  
```
scipy.stats.kstest(x, 'norm')
```
result = (statistics, pvalue)의 튜플 형태  
    pvalue가 **특정 수치 미만이면 정규성을 따른다**고 판단  

- 단일 표본 t-검정  
```
scipy.stats.ttest_1samp(x, popmean)
```
popmean = 우리가 검정하고자 하는 기준값
result = (statistics, pvalue)의 튜플 형태  
    statistics가 양수면 x의 평균이 popmean보다 큰 것이며, 음수면 x의 평균이 popmean보다 작음을 의미  
    pvalue가 **특정 수치 미만이면 x는 popmean과 같지 않다**고 판단  

- 윌콕슨 부호-순위 검정
정규성을 띠지 않을 때 하는 검정으로 popmean은 x의 중위수로 설정됨  
평균이 들어가면 평균과 평균의 차이를 분자로 두는 것이 되기 때문에 의미가 없어짐  
```
scipy.stats.wilcoxon(x)
```
result = (statistics, pvalue)의 튜플 형태  
    단일 표본 t-검정과 결과 해석이 같음

# Ch10-2. 독립 표본 t-검정 
서로 다른 두 그룹의 데이터 평균을 비교하는 가설 검정 기법이고, 단일 표본 t검정보다 많이 쓰인다.  

$ H_0: \mu_a = \mu_b $ (그룹 a의 표본 평균과 그룹 b의 표본 평균)  
$ H_1: \mu_a > \mu_b \space or \space \mu_a < \mu_b \space or \space \mu_a \not= \mu_b $  

가설 수립 예시 :  지점 a의 일별 판매량과 지점 b의 일별 판매량이 있다. 판매량 간 유의미한 차이가 있는가? (두 차이가 유의한지, 단순한 우연인지 검정)  

## 독립 표본 t-검정 선행 조건 

- 독립성: 두 그룹은 서로 독립적이어야 함  
- 정규성: 데이터는 정규분포를 따라야 함  
    - 정규성을 따르지 않으면 비모수 검정인 Mann-Whitney 검정을 수행해야 함  
- 등분산성: 두 그룹의 데이터에 대한 분산이 같아야 함  
    분산이 다르면 평균이 실제로 같아도 영향을 받을 수 있다. 두 그룹의 차이가 유의하지 않은 건 단순히 분산이 달라서 일수 있다는 것이다.  
    사실, 분산이 같을 때 할 수 있는 검정과 같지 않을 때 할 수 있는 검정이 따로 있다. 통계량만 조금씩 달라진다.  
    - Levene의 등분산 검정: p-value가 0.05 미만이면 분산이 다르다고 판단  
    - 분산이 같은지 다른지에 따라 사용하는 통계량이 달라지므로, 설정만 달리 해주면 됨  

## 독립 표본 t-검정 통계량 


*(출처: 강의 10-2 독립표본 t-검정)*

분산이 같을 때와 다를 때는 s의 구조만 다르다. 서로의 평균이 값이 같으면 0에 가까워지고, 차이가 크면 클 수록 t의 절대값이 커지는 형태라는 것!  
분자에는 평균의 차이, 분모에는 표준편차가 있다.  




## 파이썬을 이용한 독립 표본 t-검정

- 정규성 검정(KS test)  
```
scipy.stats.kstest(x, 'norm')
```
    **pvalue가 특정 수치 미만이면 정규성을 따른다**고 판단  
- 등분산성 검정(Levene test)  
```
scipy.stats.levene(s1, s2, s3, ...)
# s1, s2 는 샘플(배열)로 여러개 그룹이 동시에 들어갈 수 있음. 독립 표본 t-검정은 두 그룹만 비교하므로 둘 만들어갈 것.  
```
    **pvalue가 특정 수치 미만이면 샘플 간 분산이 같지 않다**고 판단  
- 독립 표본 t-검정  
```
scipy.stats.ttest_ind(a, b, equal_var)
# a, b: 두 그룹의 데이터(배열)
# eqaul_var: 등분산성을 만족하는지 여부(T/F)
```
    statistics가 양수면 a(첫번째 그룹)의 평균이 더 크다고 판단  
    **pvalue가 특정 수치 미만이면 a와 b의 평균이 같지 않다**고 판단  
- Mann-Whitneyu 검정  
```
scipy.stats.mannwhitneyu(a, b)
# a, b: 두 그룹의 데이터(배열)
```
    result = (statstics, pvalue)의 튜플 형태  
    **pvalue가 특정 수치 미만이면 a와 b의 평균이 같지 않다**고 판단  


````python
df1 = pd.read_csv("반별_점수_type1.csv", encoding = "cp949")
df1.head()
# 독립표본 t-검정을 할 때 보게될 데이터 구조로 나누어 type1과 type2로 나누어져있음
````

Scipy 모듈을 쓸 때는 Series보다는 ndarray를 쓰는 것이 적합하기 때문에 values로 시리즈에서 데이터만 데려온다. 

````python
group_A = df1['점수'].loc[df1['반'] == 'A'].values
group_B = df1['점수'].loc[df1['반'] == 'B'].values

print(group_A)
print(group_B)
````

````python
# kstest를 이용한 정규성 검정
# pvalue가 0.0 이므로 모두 정규 분포를 띔을 확인
print(kstest(group_A, 'norm'))
print(kstest(group_B, 'norm'))
````

````python
# 등분산 검정: 통계적으로 같은 분산을 갖는지 검정
# pvalue가 0.05 미만이 아니므로, 등분산을 띈다고 볼 수 있음
levene(group_A, group_B)
````

````python
# 분산 확인
print(np.var(group_A, None, ddof = 1))
print(np.var(group_B, None, ddof = 1))
````

````python
# 독립 표본 t-검정: 0.018로 A와 B간에는 차이가 존재가 유의미함을 확인 
# 레벤느 검정에서 등분산을 띤다고 보았으므로 equal_var = True
print(ttest_ind(group_A, group_B, equal_var = True)) 
````

통계량이 양수이므로 A반의 성적이 대체적으로 더 좋은 것으로 확인할 수 있다.
