# Ch10-1. 영가설과 대립가설, p-value

## 통계적 가설 검정
데이터를 바탕으로 통계적으로 가설의 유의미 여부를 검정하는 것을 통계적 가설 검정이라 한다.  
**가지고 있는 샘플**을 가지고 **모집단의 특성**에 대한 가설에 대한 통계적 유의성을 검정하는 일련의 과정.  
가지고 있는 데이터로 전체 데이터(모집단)을 알고 싶은 상태이다. 모집단을 정확히는 알 수 없다. 그래서 모집단의 부분집합인 가지고 있는 데이터로 추정하는 과정을 거치게 된다.  
- 수집한 데이터는 매울 특별한 경우를 제외하고는 전부 샘플이며, 모집단은 정확히 알 수 없는 경우가 더 많음  
- 통계적 유의성: 어떤 실험 결과(데이터)가 확률적으로 봐서 단순한 우연이 아니라고 판단될 정도로 의미가 있음을 뜻함. 즉 단순한 우연이라고 치부하기엔 확률이 작다, 우연이 아니다는 의미. 

통계적 가설 검정과정은 5단계이다
- 영 가설과 대립 가설 설정  
- 가설을 검정하기 위한 검정 통계량 설정  
- 가설의 유의성을 판단하기 위한 기각역 설정  
- 검정 통계량 계산  
- 가설이 유의한지 아닌지 의사 결정  



## 영 가설(귀무 가설)과 대립 가설
##### 영 가설(null hypothesis, $H_0$)
> 정의  
    - 특졀한 증거가 없으면 참으로 추정되는 가설  
    - 우리의 관심 대상이 아닌 가설(검정으로 영 가설을 기각해야 통계적 의미가 있음으로 판단)  
    - 기각하기 전까지는 참이라고 봄  
> 설정 방법  
    - 모집단에 대한 특성을 등호로 표기

##### 대립 가설(alternative hypothesis, $H_1, H_a$)
> 정의  
    - 특별한 증거가 없으면 거짓으로 추정되는 가설  
    - 우리의 관심 대상인 가설  
> 설정 방법  
    - 모집단에 대한 특성을 부등호로 표기(단측 검정, 양측 검정)  
> 양측 검정  
    - $H_0$ : n과 같다.  
    - $H_1$ : n과 같지 않다. (미만이거나 초과일 경우 둘 다 해당)  
> 단측 검정  
    - $H_0$ : n와 같다.  
    - $H_1$ : n보다 크다.  

## 오류의 구분

가설 검정에서 발생하는 오류는 두 가지로 나타난다. 귀무가설을 기준으로 참을 거짓으로, 거짓을 참으로 하여 1종과 2종으로 나눈다.  

> 1종 오류(type 1 Error)  
: 참을 거짓으로 하는 것. 참인데도 불구하고 귀무 가설이 가짜라고 하는 오류.   

> 2종 오류(type 2 Error)  
: 거짓을 참으로 하는 것. 귀무 가설이 거짓인데 참이라고 하는 오류.  


## 유의 확률, p-value

유의 확률 p-value: '영 가설이 맞다고 가정할 때 얻은 결과'와 다른 결과가 관측될 확률로, 그 값이 작을 수록 영 가설을 기각할 근거가 된다.  
보통, **p-value가 0.05 혹은 0.01 미만이면 영 가설을 기각한다.**  

예를 들어, 영 가설이 맞다고 가정하고, 데이터 관측을 했더니, 영 가설이 참일 때 나올 수 있는 데이터가 아니었다. 그럼 영 가설을 기각할 수 있다.  

주의! 영 가설을 기각한다고 해서 반드시 대립 가설이 참인 것은 아니다.  특히 단측 검정에서 그렇다. 



# Ch10-2. 단일 표본 t검정과 독립 표본 t검정

## 단일 표본 t검정
그룹의 평균이 기준 값과 차이가 있는지 확인하는 검정 방법  
어떠한 가설 검정을 공부할 때는 목적을 제대로 안 다음, 그것의 영 가설과 대립 가설을 아는 것이 중요하다. (! 그리고 구현하는 것은 어렵지 않고, 가설이 무엇이고, 검정할 때 이 방법을 쓰면 좋겠다고 생각하게 되는 것이다.)

하나의 영 가설에도 여러개의 대립가설이 생길 수 있다.  
$H_0$ : $\bar {x}  = \mu$, 평균이 기준 값과 같다.    
$H_1$ : 같지 않다, 기준 값보다 크거나, 작거나.  

### 단일 표본 t-검정의 선행 조건
- 정규 분포를 따라야 수행할 수 있음.  
    따라서 Kolmogorov-Smornov나 Shapiro-Wilk를 사용한 **정규성 검정**이 선행되어야 한다.  
- 그렇지만 샘플 수가 많을 수록 정규성을 띌 가능성이 높아지므로, 샘플 수가 부족한 경우에만 정규성 검정을 수행한 뒤, 정규성을 띄지 않는다고 판단되면, **비모수적 방법인 부호 검정(sign test)**나 **윌콕슨 부호 - 순위 검정**을 수행해야 한다.(구현 방법만 알아보기로 함)  

일반적으로 데이터가 30개 이상이면 정규성을 띌 것으로 보아도 무방할 정도이다. 

### 단일 표본 t-검정 통계량

분자에서, 표본평균과 기준 값의 차이가 크면 클 수록 통계량의 절대값이 커진다는 의미. 분모에서, 표본수가 클 수록 t의 절대값을 줄어들게 한다. 통계량의 절대값이 크면 클수록 유의 확률이 줄어드는 구조이다.  
다시 말해, 표본 평균과 기준 값의 차이가 크고, 표본수가 많을 수록 영 가설이 기각될 확률이 높아진다.  

정규분포와 유사한 t분포 상에 위치시키는 방식으로 p-value를 계산하게 된다.  
대립가설에 따라 조금씩 다르게 계산한다.  

만약 $ \bar {x} > \mu $ 혹은 $ \bar {x} < \mu $ 의 경우에는 p-value가 각각 오른쪽(p-value가 t보다 크면 대립 가설이 참이 된다), 왼쪽에 위치하고(p-value가 t보다 작으면 대립 가설이 참이 된다),  
$ \bar {x} \not= \mu$ 인 양측 검정일 경우는 각 양 꼬리에 $t$ 와 $-t$가 위치하면서 각각 p-value /2 가된다. 


### 정규성 검정 : Kolmogorov - Smornov
KS test는 관측한 샘플들이 특정 분포를 따르는지 확인하기 위한 방법이다.  
정규분포를 따르는지 검정하는 것이라기 보다, 관측한 샘플들이 어떤 분포를 따르는지 검정하는 방법이며, 특정 분포를 정규 분포로 설정하면 정규성 검정에도 쓰일 수 있다.  

우리가 가진 실제 값과, 특정 분포를 따른다면 나올 것이라 예상되는 값의 차이가 크면 클 수록 특정 분포를 따르지 않는다고 볼 수 있다. 만약 우리가 가진 값이 특정 분포를 따른다면, 그 차이는 줄어든다.
